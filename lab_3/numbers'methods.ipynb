## –¢–µ–º–∞: _–†–æ–±–æ—Ç–∞ –∑ —á–∏—Å–ª–æ–≤–∏–º–∏ –¥–∞–Ω–∏–º–∏_
### –ú–µ—Ç–∞ —Ä–æ–±–æ—Ç–∏: _–î–æ—Å–ª—ñ–¥–∏—Ç–∏ —Ä–æ–±–æ—Ç—É –∑ —á–∏—Å–ª–æ–≤–∏–º–∏ –¥–∞–Ω–∏–º–∏_

---
### –í–∏–∫–æ–Ω–∞–Ω–Ω—è —Ä–æ–±–æ—Ç–∏

```python
%pip install numpy

def homework(message):
    print("\n>>>>>", message, "\n")

import numpy as np

# –ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä
temperature = [20.1, 20.1, 20.2, 20.4]
print("–ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å–ø–∏—Å–æ–∫ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä:", temperature)

# –ú–Ω–æ–∂–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É –Ω–∞ —á–∏—Å–ª–æ
t1 = temperature * 5
print(f"–ü—Ä–∏ –º–Ω–æ–∂–µ–Ω–Ω—ñ —Å–ø–∏—Å–∫—É –Ω–∞ 5 –º–∞—î–º–æ: {t1}")
homework("–ú–Ω–æ–∂–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É –Ω–∞ —á–∏—Å–ª–æ –¥—É–±–ª—é—î —Å–∞–º —Å–ø–∏—Å–æ–∫, –∞ –Ω–µ –µ–ª–µ–º–µ–Ω—Ç–∏ –≤ –Ω—å–æ–º—É. –¶–µ –ø–æ–≤'—è–∑–∞–Ω–æ –∑ —Ç–∏–º, —è–∫ Python –æ–±—Ä–æ–±–ª—è—î –æ–ø–µ—Ä–∞—Ü—ñ—é –º–Ω–æ–∂–µ–Ω–Ω—è –¥–ª—è —Å–ø–∏—Å–∫—ñ–≤. –£ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ —Å–ø–∏—Å–æ–∫ –ø–æ–≤—Ç–æ—Ä—é—î—Ç—å—Å—è –ø'—è—Ç—å —Ä–∞–∑—ñ–≤.")

# –ú–Ω–æ–∂–µ–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫—É –Ω–∞ 5
t1 = [x * 5 for x in temperature]
print(f"–¢–µ–ø–µ—Ä –æ—Ç—Ä–∏–º–∞—î–º–æ –º–Ω–æ–∂–µ–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞: {t1}")
homework("–¢—É—Ç –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è list comprehension. –û–ø–µ—Ä–∞—Ç–æ—Ä `x * 5` –≤–∏–∫–æ–Ω—É—î –º–Ω–æ–∂–µ–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ —Å–ø–∏—Å–∫—É, —Ç–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—ñ–¥—Ä—ñ–∑–Ω—è—î—Ç—å—Å—è –≤—ñ–¥ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–≥–æ.")

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É —É NumPy –º–∞—Å–∏–≤
t2 = np.array(temperature)
print("–ú–∞—Å–∏–≤ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º NumPy:", t2)

# –ú–Ω–æ–∂–µ–Ω–Ω—è NumPy –º–∞—Å–∏–≤—É –Ω–∞ —á–∏—Å–ª–æ
t3 = t2 * 5
print("–ü—Ä–∏ –º–Ω–æ–∂–µ–Ω–Ω—ñ NumPy –º–∞—Å–∏–≤—É –Ω–∞ 5 –æ—Ç—Ä–∏–º—É—î–º–æ:", t3)
homework("NumPy –¥–æ–∑–≤–æ–ª—è—î –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ –≤–µ–∫—Ç–æ—Ä–∏–∑–æ–≤–∞–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó. –ö–æ–ª–∏ NumPy –º–∞—Å–∏–≤ –º–Ω–æ–∂–∏—Ç—å—Å—è –Ω–∞ —á–∏—Å–ª–æ, —Ü–µ –æ–∑–Ω–∞—á–∞—î, —â–æ –æ–ø–µ—Ä–∞—Ü—ñ—è –≤–∏–∫–æ–Ω—É—î—Ç—å—Å—è –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞ –º–∞—Å–∏–≤—É.")
```

**–º–µ—Ç–æ–¥ arange**

```python
import numpy as np

# –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –º–∞—Å–∏–≤—ñ–≤
start = 5
size_of_vec = 100

# –ú–∞—Å–∏–≤–∏ —á–µ—Ä–µ–∑ range —Ç–∞ np.arange
x = range(size_of_vec)
a = np.arange(size_of_vec)
print(f"–ú–∞—Å–∏–≤ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–∏–π –ø—Ä–æ—Å—Ç–∏–º range: {list(x)[:5]} —Ç–∞ –º–µ—Ç–æ–¥–æ–º arange: {a[:5]}")
homework("–í–µ—Å—å –¥—ñ–∞–ø–∞–∑–æ–Ω —á–∏—Å–µ–ª –∑ range —Ç–∞ arange:\nRange: {}\nArange: {}".format(list(x), a))

# –ó–∞–¥–∞—î–º–æ –ø–æ—á–∞—Ç–æ–∫ –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó
a = np.arange(start, size_of_vec)
print(f"–ú–æ–∂–µ–º–æ –∑–∞–¥–∞–≤–∞—Ç–∏ –ø–æ—á–∞—Ç–æ–∫ –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –≤–µ–∫—Ç–æ—Ä–∞, –ø–µ—Ä—à–∏–π –µ–ª–µ–º–µ–Ω—Ç: {a[0]}")
homework(f"–û—Å—Ç–∞–Ω–Ω—ñ–π –¥—ñ–∞–ø–∞–∑–æ–Ω —É –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–æ–º—É –º–∞—Å–∏–≤—ñ: {a[-1]}")

x = range(start, size_of_vec)
print("–¢–µ —Å–∞–º–µ –∞–ª–µ –∑ –≤–±—É–¥–æ–≤–∞–Ω–æ—é —Ñ—É–Ω–∫—Ü—ñ—î—é range:", x)
print("–í–ª–∞—Å–Ω–µ –≤–∏–≤—ñ–¥ –∑–Ω–∞—á–µ–Ω—å:", list(x)[0])
homework(f"–ß–∏—Å–ª–æ –∑ —Å–µ—Ä–µ–¥–∏–Ω–∏ –¥—ñ–∞–ø–∞–∑–æ–Ω—É: {list(x)[len(x)//2]}")

# –î—Ä–æ–±–æ–≤—ñ —á–∏—Å–ª–∞
x = np.arange(10.4)
print("–ë—É–¥–µ–º –º–∞—Ç–∏:", x)

# –ö—Ä–æ–∫ 0.5
x = np.arange(0.5, 4, 0.5)
print("–ó –∫—Ä–æ–∫–æ–º 0.5:", x)

# –ö—Ä–æ–∫ 0.08
x = np.arange(0.04, 1.84, 0.08)
print("–ó –∫—Ä–æ–∫–æ–º 0.08:", x, "\n –ó–≤–µ—Ä–Ω—ñ—Ç—å —É–≤–∞–≥—É –Ω–∞ –∑–∞–æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è")
homework("–ó–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –∫—Ä–æ–∫–æ–º 0.08: {}".format(x))

# –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –∑ –∫—Ä–æ–∫–æ–º, —â–æ –º–∞—î —á–æ—Ç–∏—Ä–∏ –Ω—É–ª—ñ –ø—ñ—Å–ª—è –∫–æ–º–∏
x = np.arange(0.0001, 1.0001, 0.0001)
print("–ó –∫—Ä–æ–∫–æ–º 0.0001:", x)
homework("–ó–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –∫—Ä–æ–∫–æ–º 0.0001: {}".format(x))

# –û–ø–µ—Ä–∞—Ü—ñ—ó –∑ –æ—Ç—Ä–∏–º–∞–Ω–∏–º–∏ —á–∏—Å–ª–∞–º–∏
y = x * 2.5
print("–ü—ñ—Å–ª—è –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–Ω—è:", y)
homework(f"–ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ —Å—Ç–µ–ø–µ–Ω—è (y**2) —Ç–∞ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0.4:\n{y**2 / 0.4}")
```

##### **–ú–µ—Ç–æ–¥ linspace**

```python
import numpy as np

# –ó–≥–µ–Ω–µ—Ä—É–≤–∞—Ç–∏ —á–∏—Å–ª–æ–≤–∏–π –¥—ñ–∞–ø–∞–∑–æ–Ω
start = 5
size_of_vec = 100

# –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º linspace
a = np.linspace(start, size_of_vec)
print(f"–ó–≥–µ–Ω–µ—Ä—É–≤–∞–ª–∞ {len(a)} —á–∏—Å–µ–ª –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ {a[0]} –¥–æ {a[-1]}")

# –í–∏–≤–µ—Å—Ç–∏ –∫–æ–∂–Ω–µ –¥—Ä—É–≥–µ —á–∏—Å–ª–æ –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–æ–≥–æ –¥—ñ–∞–ø–∞–∑–æ–Ω—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ü–∏–∫–ª—É
homework("–ö–æ–∂–Ω–µ –¥—Ä—É–≥–µ —á–∏—Å–ª–æ:")
for i in range(1, len(a), 2):  # –∫—Ä–æ–∫ 2
    print(a[i], end=" ")

print("\n\n–ú–æ–∂–Ω–∞ –∑–º–µ–Ω—à–∏—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª –¥–æ 5:")
a = np.linspace(start, size_of_vec, 5)
print(a)
homework("–ó–±—ñ–ª—å—à—É—é –∫—ñ–ª—å–∫—ñ—Å—Ç—å —á–∏—Å–µ–ª –¥–æ 100 —ñ –≤–∏–≤–µ–¥—É –ø–µ—Ä—à—ñ —Ç–∞ –æ—Å—Ç–∞–Ω–Ω—ñ 10:")
a = np.linspace(start, size_of_vec, 100)
print(f"–ü–µ—Ä—à—ñ 10 —á–∏—Å–µ–ª: {a[:10]}")
print(f"–û—Å—Ç–∞–Ω–Ω—ñ 10 —á–∏—Å–µ–ª: {a[-10:]}")

# –í–∏–∫–ª—é—á–µ–Ω–Ω—è –æ—Å—Ç–∞–Ω–Ω—å–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è
print("""
–ü–µ—Ä—à–µ —Ç–∞ –æ—Å—Ç–∞–Ω–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∑–∞–≤–∂–¥–∏ –≤—Ö–æ–¥—è—Ç—å —É –∑–≥–µ–Ω–µ—Ä–æ–≤–∞–Ω–∏–π –¥—ñ–∞–ø–∞–∑–æ–Ω. 
–í—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –ø—Ä–∏ —Ü—å–æ–º—É –≤–∏—Ä–∞—Ö–æ–≤—É—î—Ç—å—Å—è –∫—Ä–æ–∫.
–¶–µ –º–æ–∂–Ω–∞ –∑–º—ñ–Ω–∏—Ç–∏, –≤–∏–∫–ª—é—á–∏–≤—à–∏ –æ—Å—Ç–∞–Ω–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –∑ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥—ñ–∞–ø–∞–∑–æ–Ω—É.
""")
print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("–ö—Ä–æ–∫ —á–∏—Å–ª–æ–≤–æ–≥–æ —Ä—è–¥—É, –∫–æ–ª–∏ –æ—Å—Ç–∞–Ω–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏–∫–ª—é—á–µ–Ω–æ:")
step = (size_of_vec - start) / 5
print(f"–ö—Ä–æ–∫ (–≤—Ä—É—á–Ω—É): {step}")

# –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ–≥–æ –≤–∏–∑–Ω–∞—á–µ–Ω–Ω—è –∫—Ä–æ–∫—É
_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(f"–î–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è, –≤–∏–∑–Ω–∞—á—É –∫—Ä–æ–∫ –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ, —É –Ω–∞—à–æ–º—É –≤–∏–ø–∞–¥–∫—É —Ü–µ {sp_1} (–∑ endpoint=True) —Ç–∞ {sp_2} (–∑ endpoint=False)")

homework(f"–ö—Ä–æ–∫ –∑ endpoint=True: {sp_1}, –∫—Ä–æ–∫ –∑ endpoint=False: {sp_2}. –†—ñ–∑–Ω–∏—Ü—è –ø–æ—è—Å–Ω—é—î—Ç—å—Å—è —Ç–∏–º, —â–æ –∑ `endpoint=False` –æ—Å—Ç–∞–Ω–Ω—î —á–∏—Å–ª–æ –≤–∏–∫–ª—é—á–µ–Ω–æ –∑ —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É, —Ç–æ–º—É –∫—Ä–æ–∫ –±—ñ–ª—å—à–∏–π.")
```

##### **–¢–∏–ø–∏ –º–∞—Å–∏–≤—ñ–≤/–≤–µ–∫—Ç–æ—Ä—ñ–≤**

```python
import numpy as np

# –ü—Ä–∏–∫–ª–∞–¥–∏ –º–∞—Å–∏–≤—ñ–≤ —Ä—ñ–∑–Ω–æ—ó —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ
examples = {
    '–°–∫–∞–ª—è—Ä': np.array(1), 
    '–í–µ–∫—Ç–æ—Ä –†—è–¥–æ–∫': np.array([1, 2, 3]), 
    '–í–µ–∫—Ç–æ—Ä –°—Ç–æ–≤–ø–µ—Ü—å': np.array([[1], [2], [3]]), 
    '–ú–∞—Ç—Ä–∏—Ü—è': np.array([[1, 2], [3, 4], [5, 6]]),
    '–ë–∞–≥–∞—Ç–æ–≤–∏–º—ñ—Ä–Ω–∞ –ú–∞—Ç—Ä–∏—Ü—è': np.array([[[111, 112], [121, 122]],
                                        [[211, 212], [221, 222]],
                                        [[311, 312], [321, 322]]])
}

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –ø—Ä–∏–∫–ª–∞–¥—É
for key in examples:
    print(f"–†–æ–∑–≥–ª—è–¥–∞—î–º–æ –ø—Ä–∏–∫–ª–∞–¥ –∑ {key}. \n–ó–Ω–∞—á–µ–Ω–Ω—è x –±—É–¥–µ: \n", examples[key])
    print(f"–¢–∏–ø {key} x: ", type(examples[key]))
    print(f"–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å/–≤–∏–º—ñ—Ä {key} x:", np.ndim(examples[key]))
    print(f"–§–æ—Ä–º–∞ {key} x: ", examples[key].shape)
    print(f"–ö—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Ö: {np.size(examples[key])}")
    print("-=-" * 10)

# –ü–æ—è—Å–Ω–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É
homework("""–§—É–Ω–∫—Ü—ñ—è np.ndim –ø–æ–≤–µ—Ä—Ç–∞—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–º—ñ—Ä—ñ–≤ –º–∞—Å–∏–≤—É (rank). 
–í–µ–∫—Ç–æ—Ä-—Ä—è–¥–æ–∫ —Ç–∞ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø–µ—Ü—å –º–∞—é—Ç—å –æ–¥–Ω–∞–∫–æ–≤—É —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å 1D (–æ–¥–∏–Ω –≤–∏–º—ñ—Ä), —Ç–æ–º—É np.ndim –ø–æ–≤–µ—Ä—Ç–∞—î 1 –¥–ª—è –æ–±–æ—Ö. 
–§–æ—Ä–º–∞ (shape) –≤–∫–∞–∑—É—î –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É –∫–æ–∂–Ω–æ–º—É –≤–∏–º—ñ—Ä—ñ: –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, (3,) –¥–ª—è —Ä—è–¥–∫–∞ —Ç–∞ (3, 1) –¥–ª—è —Å—Ç–æ–≤–ø—Ü—è. 
–¶–µ –∑–Ω–∞—á–µ–Ω–Ω—è shape –ø–æ—è—Å–Ω—é—î –≤—ñ–¥–º—ñ–Ω–Ω—ñ—Å—Ç—å –º—ñ–∂ –≤–µ–∫—Ç–æ—Ä–æ–º-—Ä—è–¥–∫–æ–º —ñ –≤–µ–∫—Ç–æ—Ä–æ–º-—Å—Ç–æ–≤–ø—Ü–µ–º.""")

# –ó–∞–≤–¥–∞–Ω–Ω—è: –ó–∞–¥–∞–π—Ç–µ –≤–ª–∞—Å–Ω—ñ –≤–µ–∫—Ç–æ—Ä–∏
my_vector_row = np.array([10, 20, 30, 40])
my_vector_column = np.array([[10], [20], [30], [40]])

print("–ú—ñ–π –≤–µ–∫—Ç–æ—Ä-—Ä—è–¥–æ–∫:\n", my_vector_row)
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å:", np.ndim(my_vector_row))
print("–§–æ—Ä–º–∞:", my_vector_row.shape)

print("–ú—ñ–π –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø–µ—Ü—å:\n", my_vector_column)
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å:", np.ndim(my_vector_column))
print("–§–æ—Ä–º–∞:", my_vector_column.shape)
```

```python
import numpy as np

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–∞—Å–∏–≤—É
A = np.arange(50)

# –£–º–æ–≤–∞ –¥–ª—è –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª
all_even = (A % 2 == 0)

# –ö–æ–º–ø–ª–µ–∫—Å–Ω–∞ —É–º–æ–≤–∞ –¥–ª—è –≤–∏–±–æ—Ä—É —á–∏—Å–µ–ª
complex_condition = (A < 30) & (A / 5 < 2)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ñ–≤
print(f"""–°–ø–æ—á–∞—Ç–∫—É –ø–æ—Ç—Ä—ñ–±–Ω–æ —Å—Ç–≤–æ—Ä–∏—Ç–∏ —É–º–æ–≤—É –∑–∞ —è–∫–æ—é –∑–¥—ñ–π—Å–Ω—é–≤–∞—Ç–∏ –≤–∏–±—ñ—Ä–∫—É.
–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–æ–≤–∏ –±—É–¥–µ –∞–±–æ {all_even[0]} –¥–ª—è {A[0]} –∞–±–æ {all_even[1]} –¥–ª—è {A[1]}.
–ü—ñ—Å–ª—è –ø—ñ–¥—Å—Ç–∞–≤–ª—è–Ω–Ω—è —É–º–æ–≤–∏ –ø–æ—à—É–∫—É –ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª –≤ –º–∞—Å–∏–≤ –æ—Ç—Ä–∏–º–∞—î–º–æ:
{A[all_even]}
–ê–±–æ –∑–∞—Å—Ç–æ—Å—É–≤–∞–≤—Ç–∏ –±—ñ–ª—å—à –∫–æ–º–ø–ª–µ–∫—Å–Ω—É —É–º–æ–≤—É –∑ –¥–≤–æ—Ö –ø–æ—Ä—ñ–≤–Ω—è–Ω—å:
{A[complex_condition]}
""")

# –ó–∞–≤–¥–∞–Ω–Ω—è: –í–∏–≤–µ—Å—Ç–∏ –≤—Å—ñ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑ –º–∞—Å–∏–≤—É
homework("–í–∏–≤–µ–¥–∂—É –≤—Å—ñ –Ω–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –∑ –¥–∞–Ω–æ–≥–æ –º–∞—Å–∏–≤—É.")

# –£–º–æ–≤–∞ –¥–ª—è –Ω–µ–ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª (—è–∫—ñ –Ω–µ —î –ø–∞—Ä–Ω–∏–º–∏)
all_odd = (A % 2 != 0)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –Ω–µ–ø–∞—Ä–Ω–∏—Ö —á–∏—Å–µ–ª
print(f"–ù–µ–ø–∞—Ä–Ω—ñ —á–∏—Å–ª–∞ –≤ –º–∞—Å–∏–≤—ñ: {A[all_odd]}")
```

##### **–§—É–Ω–∫—Ü—ñ—ó –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å: ones, zeros, identity, eye, empty**

```python
import numpy as np  # –Ü–º–ø–æ—Ä—Ç—É—î–º–æ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É NumPy –¥–ª—è —Ä–æ–±–æ—Ç–∏ –∑ –º–∞—Å–∏–≤–∞–º–∏

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ –∑ —É—Å—ñ—Ö –æ–¥–∏–Ω–∏—Ü—å
o = np.ones(5)  # –ú–∞—Å–∏–≤ –∑ –ø'—è—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç –¥–æ—Ä—ñ–≤–Ω—é—î 1

# –°—Ç–≤–æ—Ä—é—î–º–æ –º–∞—Å–∏–≤ –∑ —É—Å—ñ—Ö –Ω—É–ª—ñ–≤
z = np.zeros(5)  # –ú–∞—Å–∏–≤ –∑ –ø'—è—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç –¥–æ—Ä—ñ–≤–Ω—é—î 0

# –°—Ç–≤–æ—Ä—é—î–º–æ –æ–¥–∏–Ω–∏—á–Ω—É –º–∞—Ç—Ä–∏—Ü—é 5x5
i = np.identity(5)  # –û–¥–∏–Ω–∏—á–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è —Ä–æ–∑–º—ñ—Ä—É 5x5 (–¥—ñ–∞–≥–æ–Ω–∞–ª—å –≤—Å—ñ 1, —ñ–Ω—à—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ 0)

# –°—Ç–≤–æ—Ä—é—î–º–æ –æ–¥–∏–Ω–∏—á–Ω—É –º–∞—Ç—Ä–∏—Ü—é 5x5 –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —ñ–Ω—à–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó
e = np.eye(5)  # –¢–µ–∂ —Å—Ç–≤–æ—Ä—é—î –æ–¥–∏–Ω–∏—á–Ω—É –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä—É 5x5

# –°—Ç–≤–æ—Ä—é—î–º–æ "–ø—É—Å—Ç–∏–π" –º–∞—Å–∏–≤, –∑–Ω–∞—á–µ–Ω–Ω—è –≤ —è–∫–æ–º—É –∑–∞–ª–µ–∂–∞—Ç—å –≤—ñ–¥ –ø–∞–º'—è—Ç—ñ
t = np.empty(5)  # –ú–∞—Å–∏–≤ –∑ –ø'—è—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, —è–∫–∏–π –Ω–µ —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑—É—î—Ç—å—Å—è, —ñ –∑–Ω–∞—á–µ–Ω–Ω—è –≤ –Ω—å–æ–º—É –≤–∏–ø–∞–¥–∫–æ–≤—ñ

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –º–∞—Å–∏–≤—ñ–≤ —ñ —ó—Ö –ø–æ—è—Å–Ω–µ–Ω–Ω—è:
print(f"""
–í—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ 1:
{o}  # –í–∏–≤–æ–¥–∏–º–æ –º–∞—Å–∏–≤ –∑ –æ–¥–∏–Ω–∏—Ü—è–º–∏

–í—Å—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ 0:
{z}  # –í–∏–≤–æ–¥–∏–º–æ –º–∞—Å–∏–≤ –∑ –Ω—É–ª—è–º–∏

–ú–∞—Ç—Ä–∏—Ü—è, –≤ —è–∫–æ—ó –¥—ñ–∞–≥–æ–Ω–∞–ª—å –≤—Å—ñ 1:
{i}  # –í–∏–≤–æ–¥–∏–º–æ –æ–¥–∏–Ω–∏—á–Ω—É –º–∞—Ç—Ä–∏—Ü—é —Ä–æ–∑–º—ñ—Ä—É 5x5

–¢–∞–∫–æ–∂, –¥—ñ–∞–≥–æ–Ω–∞–ª—å –≤—Å—ñ 1 (–∑–º–µ–Ω—à–∏–º–æ –≤–∏–≤—ñ–¥):
{e[:2, :2]}  # –í–∏–≤–æ–¥–∏–º–æ –≤–µ—Ä—Ö–Ω—ñ–π –ª—ñ–≤–∏–π –ø—ñ–¥–º–∞—Å–∏–≤ —Ä–æ–∑–º—ñ—Ä—É 2x2 –∑ –æ–¥–∏–Ω–∏—á–Ω–æ—ó –º–∞—Ç—Ä–∏—Ü—ñ (–¥—ñ–∞–≥–æ–Ω–∞–ª—å–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏)

–ü—É—Å—Ç–∞ –º–∞—Ç—Ä–∏—Ü—è –±—É–¥–µ –º—ñ—Å—Ç–∏—Ç–∏ –Ω–∞–±—ñ—Ä –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö —á–∏—Å–µ–ª, —è–∫—ñ –∑–∞–ø–∏—Å–∞–Ω—ñ –≤ –ø–∞–º'—è—Ç—ñ, –¥–µ –±—É–≤ —ñ–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–æ–≤–∞–Ω–∏–π –º–∞—Å–∏–≤:
{t}  # –í–∏–≤–æ–¥–∏–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è –∑ "–ø—É—Å—Ç–æ–≥–æ" –º–∞—Å–∏–≤—É, —â–æ –∑–∞–ª–µ–∂–∏—Ç—å –≤—ñ–¥ –ø–∞–º'—è—Ç—ñ
""")
```

1. –ú–µ—Ç–æ–¥ arange
- :question: –ú–µ—Ç–æ–¥ arange –≥–µ–Ω–µ—Ä—É—î —á–∏—Å–ª–æ–≤—É –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å –Ω–∞ –æ—Å–Ω–æ–≤—ñ –ø–æ—á–∞—Ç–∫–æ–≤–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è, –∫—ñ–Ω—Ü–µ–≤–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è —Ç–∞ –∫—Ä–æ–∫—É
> –§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ü—å–æ–≥–æ –≤–∏–≥–ª—è–¥–∞—î —Ç–∞–∫: a(n)=a(0)+n*d
- :question: –§–æ—Ä–º—É–ª–∞ –≤–∫–∞–∑—É—î, —â–æ –Ω–∞—Å—Ç—É–ø–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –ø–æ—Å–ª—ñ–¥–æ–≤–Ω–æ—Å—Ç—ñ –≥–µ–Ω–µ—Ä—É—î—Ç—å—Å—è —à–ª—è—Ö–æ–º –¥–æ–¥–∞–≤–∞–Ω–Ω—è –∫—Ä–æ–∫—É –¥–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç–∞.

```python
import numpy as np
a = np.arange(1, 10, 2)
print(a)
```

2. –ú–µ—Ç–æ–¥ linspace
- :question: –ú–µ—Ç–æ–¥ linspace –≥–µ–Ω–µ—Ä—É—î —á–∏—Å–ª–æ–≤—É –ø–æ—Å–ª—ñ–¥–æ–≤–Ω—ñ—Å—Ç—å, –≤–∫–∞–∑—É—é—á–∏ –ø–æ—á–∞—Ç–∫–æ–≤–µ —Ç–∞ –∫—ñ–Ω—Ü–µ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è, –∞ —Ç–∞–∫–æ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤. –í—ñ–Ω –∞–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –æ–±—á–∏—Å–ª—é—î –∫—Ä–æ–∫ –¥–ª—è —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ–≥–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—É –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º—ñ–∂ –ø–æ—á–∞—Ç–∫–æ–º —ñ –∫—ñ–Ω—Ü–µ–º. 
> –§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ü—å–æ–≥–æ –≤–∏–≥–ª—è–¥–∞—î —Ç–∞–∫: a(n)=(a(0)+a(T)-a(0)/N-1)*n; –¥–µ - a(0)‚Äî –ø–æ—á–∞—Ç–∫–æ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è, ùëé(–¢)‚Äî –∫—ñ–Ω—Ü–µ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è, N‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, n ‚Äî —ñ–Ω–¥–µ–∫—Å –µ–ª–µ–º–µ–Ω—Ç–∞ (–≤—ñ–¥ 0 –¥–æ N‚àí1).
- :question: –§–æ—Ä–º—É–ª–∞ –≤–∫–∞–∑—É—î, —â–æ –∫—Ä–æ–∫ –æ–±—á–∏—Å–ª—é—î—Ç—å—Å—è —Ç–∞–∫, —â–æ–± —á–∏—Å–ª–∞ —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ –∑–∞–ø–æ–≤–Ω–∏–ª–∏ –≤–µ—Å—å –¥—ñ–∞–ø–∞–∑–æ–Ω –≤—ñ–¥ a(0) –¥–æ –∞(–¢)

**–ü—Ä–∏–∫–ª–∞–¥**

```python
import numpy as np
b = np.linspace(1, 10, 5)
print(b)
```

### –ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —Ñ–æ—Ä–º—É–ª —ñ –≤–∏–ø–∞–¥–∫—ñ–≤ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è

> –§–æ—Ä–º—É–ª–∏:

> - `linspace` –ø—ñ–¥—Ö–æ–¥–∏—Ç—å –¥–ª—è —Å–∏—Ç—É–∞—Ü—ñ–π, –∫–æ–ª–∏ –ø–æ—Ç—Ä—ñ–±–Ω–æ —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ —Ä–æ–∑–ø–æ–¥—ñ–ª–∏—Ç–∏ –∑–∞–¥–∞–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –º—ñ–∂ –ø–æ—á–∞—Ç–∫–æ–≤–∏–º —ñ –∫—ñ–Ω—Ü–µ–≤–∏–º –∑–Ω–∞—á–µ–Ω–Ω—è–º, –≥–∞—Ä–∞–Ω—Ç—É—é—á–∏, —â–æ –≤—Å—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –±—É–¥—É—Ç—å —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ —Ä–æ–∑–ø–æ–¥—ñ–ª–µ–Ω—ñ.

### –°—Ç–≤–æ—Ä—é—é –¥–µ–∫—ñ–ª—å–∫–∞ –ø—Ä–æ—Å—Ç–∏—Ö –í–µ–∫—Ç–æ—Ä—ñ–≤ —Ç–∞ –ú–∞—Ç—Ä–∏—Ü—å —Ç–∞ –≤–∏–≤–µ–¥–∂—É —ó—Ö —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å. –í–∏–∫–æ–Ω—É—é –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –Ω–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–∏–º–∏ –í–µ–∫—Ç–æ—Ä–∞–º/–ú–∞—Ç—Ä–∏—Ü—è–º–∏;

**1. –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –ø—Ä–æ—Å—Ç–∏—Ö –≤–µ–∫—Ç–æ—Ä—ñ–≤ —Ç–∞ –º–∞—Ç—Ä–∏—Ü—å**

```python
import numpy as np

# –°—Ç–≤–æ—Ä—é—é –≤–µ–∫—Ç–æ—Ä-—Ä—è–¥–æ–∫ —Ç–∞ –≤–µ–∫—Ç–æ—Ä-—Å—Ç–æ–≤–ø–µ—Ü—å
vector_row = np.array([1, 2, 3, 4, 5])
vector_column = np.array([[1], [2], [3], [4], [5]])

# –°—Ç–≤–æ—Ä—é—é –º–∞—Ç—Ä–∏—Ü—é 2x3
matrix_2x3 = np.array([[1, 2, 3], [4, 5, 6]])

# –°—Ç–≤–æ—Ä—é—é –º–∞—Ç—Ä–∏—Ü—é 3x3
matrix_3x3 = np.array([[7, 8, 9], [10, 11, 12], [13, 14, 15]])

# –í–∏–≤–æ–¥–∂—É —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–µ–∫—Ç–æ—Ä–∞-—Ä—è–¥–∫–∞:", vector_row.shape)
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –≤–µ–∫—Ç–æ—Ä–∞-—Å—Ç–æ–≤–ø—Ü—è:", vector_column.shape)
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü—ñ 2x3:", matrix_2x3.shape)
print("–†–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü—ñ 3x3:", matrix_3x3.shape)
```

**2. –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω—ñ –æ–ø–µ—Ä–∞—Ü—ñ—ó –Ω–∞–¥ –≤–µ–∫—Ç–æ—Ä–∞–º–∏ —Ç–∞ –º–∞—Ç—Ä–∏—Ü—è–º–∏**
> –û–ø–µ—Ä–∞—Ü—ñ—ó –∑ –≤–µ–∫—Ç–æ—Ä–∞–º–∏

```python
# –î–æ–¥–∞–≤–∞–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä—ñ–≤
vector_add = vector_row + np.array([1, 1, 1, 1, 1])
print("\n–î–æ–¥–∞–≤–∞–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä—ñ–≤:\n", vector_add)

# –ú–Ω–æ–∂–µ–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä–∞ –Ω–∞ —Å–∫–∞–ª—è—Ä
vector_multiply = vector_row * 2
print("\n–ú–Ω–æ–∂–µ–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä–∞ –Ω–∞ —Å–∫–∞–ª—è—Ä:\n", vector_multiply)

# –í—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä—ñ–≤
vector_subtract = vector_row - np.array([0, 1, 2, 3, 4])
print("\n–í—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä—ñ–≤:\n", vector_subtract)
```

> –û–ø–µ—Ä–∞—Ü—ñ—ó –∑ –º–∞—Ç—Ä–∏—Ü—è–º–∏

```python
# –î–æ–¥–∞–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å
matrix_add = matrix_3x3 + np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
print("\n–î–æ–¥–∞–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å:\n", matrix_add)

# –ú–Ω–æ–∂–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ —Å–∫–∞–ª—è—Ä
matrix_multiply = matrix_3x3 * 3
print("\n–ú–Ω–æ–∂–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ –Ω–∞ —Å–∫–∞–ª—è—Ä:\n", matrix_multiply)

# –¢—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ
matrix_transpose = matrix_3x3.T
print("\n–¢—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—ñ:\n", matrix_transpose)
```
